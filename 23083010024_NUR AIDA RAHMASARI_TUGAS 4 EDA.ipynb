### Outlier

#### Menggunakan Boxplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('D:/SEMESTER 3/EDA/data_dummy_dd_baru.csv')

plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age']])
plt.xticks(rotation=45)
plt.show()
```

#### Metode Z-Score

```python
import numpy as np
from scipy.stats import zscore

z_scores = np.abs(zscore(df.select_dtypes(include=['float64', 'int64']))) 

threshold = 3
outliers = np.where(z_scores > threshold)

print("Outliers ditemukan pada indeks:", outliers)
```

#### Menggunakan IQR

```python
Q1 = df.select_dtypes(include=['float64', 'int64']).quantile(0.25)
Q3 = df.select_dtypes(include=['float64', 'int64']).quantile(0.75)
IQR = Q3 - Q1

outliers = ((df.select_dtypes(include=['float64', 'int64']) < (Q1 - 1.5 * IQR)) | 
            (df.select_dtypes(include=['float64', 'int64']) > (Q3 + 1.5 * IQR)))

print(outliers)
```

Saya akan menginterpretasikan beberapa hasil dari cek outliers menggunakan metode Z-Score, Baris ke-4 memiliki outlier di kolom ke-6, yang berarti ada nilai outlier pada kolom DiabetesPedigreeFunction untuk baris ke-4.Baris ke-7 memiliki outlier di kolom ke-2, yang berarti ada nilai outlier pada kolom BloodPressure untuk baris ke-7.Baris ke-13 memiliki outlier di kolom ke-4, yang berarti ada nilai outlier pada kolom Insulin untuk baris ke-13.

### Mengatasi Outliers

```python
import numpy as np
from scipy.stats import zscore

z_scores = np.abs(zscore(df.select_dtypes(include=['float64', 'int64'])))

threshold = 3

df_clean = df[(z_scores < threshold).all(axis=1)]

print("Data setelah menghapus outliers:")
print(df_clean)
```

Data sudah bersih dari outliers

#### Kategorisasi Variabel

```python
# Kategorisasi untuk variabel Age dan BMI
age_bins = [0, 25, 50, 100]
age_labels = ['Young', 'Middle-aged', 'Old']
df['Age_Category'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

df['BMI_Category'] = pd.qcut(df['BMI'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

print(df[['Age', 'Age_Category', 'BMI', 'BMI_Category']].head())
```

Secara keseluruhan, distribusi usia dan BMI menunjukkan beragam pola dalam data, di mana beberapa individu yang tergolong usia paruh baya memiliki BMI rendah hingga sangat tinggi, sementara individu yang lebih muda cenderung memiliki BMI yang lebih sedang.

### Transformasi data, lalu visualkan sebelum dan sesudah

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('D:/SEMESTER 3/EDA/data_dummy_dd_baru.csv')

# tanpa 'Outcome' karena itu adalah variabel target
numeric_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# 1. Visualisasi Sebelum Transformasi 
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i) 
    sns.histplot(df[column], kde=True) 
    plt.title(f'{column} - Before Scaling')
plt.tight_layout()
plt.show()

# 2. Transformasi Data dengan StandardScaler
scaler = StandardScaler()
df_scaled = df.copy()  
df_scaled[numeric_columns] = scaler.fit_transform(df[numeric_columns]) 

# 3. Visualisasi Setelah Transformasi 
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df_scaled[column], kde=True)
    plt.title(f'{column} - After Scaling')
plt.tight_layout()
plt.show()
```

Visualisasi Sebelum Transformasi:
Menggunakan histogram dan KDE (Kernel Density Estimation) untuk melihat distribusi data sebelum transformasi.

Transformasi dengan StandardScaler:
StandardScaler akan mengubah data sehingga memiliki mean 0 dan standar deviasi 1, yang bagus untuk algoritma berbasis jarak seperti KNN, SVM, dll.

Visualisasi Setelah Transformasi:
Setelah data di-transform, histogram baru ditampilkan untuk melihat bagaimana distribusi data berubah.

Sebelum transformasi: Data mungkin memiliki distribusi yang tidak normal atau berada dalam skala yang berbeda.

Setelah transformasi: Data akan lebih terpusat di sekitar 0, dan skala datanya akan seragam.
